```python
# Deciding how to fill in missing ages
for sex in ['male', 'female']:
    for Pclass in [1,2,3]:
        print titanic[(titanic['Sex']==sex) & (titanic['Pclass']==Pclass)]['Age'].mean()
```

```python
def fill_ages(titanic):
    #fills in missing ages and missing Fare
    for pclass in [1,2,3]:
        titanic.loc[titanic['Pclass']==pclass,'Fare'] = titanic[titanic['Pclass']==pclass]['Fare'].fillna(titanic[titanic['Pclass']==pclass]['Fare'].mean())
        for gender in ['male', 'female']:
            gender_class = (titanic['Sex']==gender) & (titanic['Pclass']==pclass)
            av_age = titanic[gender_class]['Age'].mean()
            titanic.loc[gender_class,'Age'] = titanic[gender_class]['Age'].fillna(av_age)
    return titanic
```

```python
titanic = fill_ages(titanic)
titanic.info()
```

```python
def cabin_dummies(titanic, cabs):
    for cab, letter in cabs:
        titanic.loc[(titanic['Cabin'].notnull() & titanic['Cabin'].str.contains(letter)),cab]=1
        titanic[cab] = titanic[cab].fillna(0)
    return titanic
```

```python
def munge(titanic):
    titanic = fill_ages(titanic)
    titanic = cabin_dummies(titanic,[['cab_A', 'A'],['cab_B','B'],['cab_C','C'],['cab_D', 'D'],['cab_E','E'],['cab_F', 'F'],['cab_G','G']])    
    titanic['female'] = titanic['Sex'].map( {'female': 1, 'male': 0} ).astype(int)
        
    class_dummies = pd.core.reshape.get_dummies(titanic['Pclass'])
    titanic[['1st_class','2nd_class']] = class_dummies[[1,2]]

    embark_dummies = pd.core.reshape.get_dummies(titanic['Embarked'])
    titanic[['Cherbourg','Queenstown']] = embark_dummies[['C','Q']]
    titanic.loc[titanic['Embarked'].isnull(),'Cherbourg']=titanic['Cherbourg'].mean()
    titanic.loc[titanic['Embarked'].isnull(),'Queenstown']=titanic['Queenstown'].mean()
    
    titanic = titanic.drop(['Embarked','Pclass', 'Sex','Cabin','Ticket','Name'],axis=1)
    return titanic
```

```python
titanic = pd.read_csv('train.csv')
titanic = munge(titanic)
titanic.info()
```

```python
def add_features(titanic):    
    titanic['female_age'] = titanic['female'] * titanic['Age']
    titanic['female_1st_class'] = titanic['female'] * titanic['1st_class']
    titanic['female_2nd_class'] = titanic['female'] * titanic['2nd_class']
    
    titanic.loc[((titanic['SibSp'] + titanic['Parch'])>0),'family'] = 1
    titanic['family'] = titanic['family'].fillna(0)

    return titanic
```

```python
titanic = add_features(titanic)
```

```python
def titanic_model(titanic, features):
    model_lr = LogisticRegression(C=1).fit
    y = titanic['Survived']
    X = titanic[features]
    X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.20,random_state=0)
    model = model_lr(X_train,y_train)
    forest_model = RandomForestClassifier(random_state=0).fit(X_train,y_train)
    print str(features) + " model:"
    print "train score: " + str(model.score(X_train,y_train))
    print "test score : " + str(model.score(X_test,y_test))
#     print "coefficients: " + str(model.coef_)
#     print "intercept: " + str(model.intercept_[0])
    print "forest score : " + str(forest_model.score(X_test,y_test))

#     print model.predict(x)
    print ""
```

```python
titanic_model(titanic, ['female'])
titanic_model(titanic,['Age','female'])
titanic_model(titanic,['female', '1st_class', '2nd_class'])
titanic_model(titanic,['female', '1st_class','2nd_class','Fare','SibSp', 'Parch'])
titanic_model(titanic,['Age','female', '1st_class','2nd_class','Fare','SibSp', 'Parch'])
titanic_model(titanic,['Age','female', '1st_class','2nd_class','Fare','SibSp', 'Parch','Queenstown','Cherbourg'])
titanic_model(titanic,['Age','female', '1st_class','2nd_class','Fare','SibSp', 'Parch','Queenstown','Cherbourg', 'female_age'])
titanic_model(titanic,['Age','female', '1st_class','2nd_class','Fare','SibSp', 'Parch','Queenstown','Cherbourg', 'female_1st_class', 'female_2nd_class'])
titanic_model(titanic,['Age','female', '1st_class','2nd_class','Fare','SibSp', 'Parch','Queenstown','Cherbourg', 'female_age', 'female_1st_class', 'female_2nd_class'])
titanic_model(titanic,['Age','female', '1st_class','2nd_class','Fare','SibSp', 'Parch','Queenstown','Cherbourg', 'female_age', 'female_1st_class', 'female_2nd_class', 'family'])
titanic_model(titanic,['Age','female', '1st_class','2nd_class','Fare','SibSp', 'Parch','Queenstown','Cherbourg', 'female_age', 'female_1st_class', 'female_2nd_class', 'family', 'cab_A', 'cab_B', 'cab_C', 'cab_D', 'cab_E', 'cab_F', 'cab_G'])
```

### Cross Validation

```python
def titanic_folds(titanic,k):  
    kf = KFold(len(titanic['Survived']), n_folds=k, shuffle=True, random_state = 0)
    avg_train_score = 0
    avg_test_score = 0
    model_lr = LogisticRegression(C=1).fit
    for train, test in kf:
        y = titanic.loc[train,'Survived']
        x = titanic.loc[train,features]
        y_test = titanic.loc[test,'Survived']
        x_test = titanic.loc[test,features]
        model = model_lr(x,y)
        avg_train_score += model.score(x,y) /k
        avg_test_score += model.score(x_test,y_test) / k

    print "k-folds = " + str(k)
    print "avg_train_score: " + str(avg_train_score)
    print "avg_test_score: " + str(avg_test_score)
    print " "
```

```python
features = ['Age','female', '1st_class','2nd_class','Fare','SibSp', 'Parch','Queenstown','Cherbourg', 'female_age', 'female_1st_class', 'female_2nd_class', 'family', 'cab_A', 'cab_B', 'cab_C', 'cab_D', 'cab_E', 'cab_F', 'cab_G']
for k in [2,3,5,7,10,15,20,30,50]:
    titanic_folds(titanic,k)
```

## ROC Curve

```python
features = ['Age','female', '1st_class','2nd_class','Fare','SibSp', 'Parch','Queenstown','Cherbourg', 'female_age', 'female_1st_class', 'female_2nd_class', 'family', 'cab_A','cab_B', 'cab_C', 'cab_D', 'cab_E', 'cab_F', 'cab_G']
x = titanic.loc[:,features]
y = titanic.loc[:,'Survived']
model_lr = LogisticRegression(C=1).fit
model = model_lr(x,y)
print model.score(x,y)
# target_predicted_proba = model_lr.predict_proba(x)
predictions = model.predict_proba(x)
```

```python
def plot_roc_curve(target_test, target_predicted_proba):
    fpr, tpr, thresholds = roc_curve(target_test, target_predicted_proba[:, 1
                                                                         ])
    roc_auc = auc(fpr, tpr)
    
    # Plot ROC curve
    plt.plot(fpr, tpr, label='ROC curve (area = %0.3f)' % roc_auc)
    plt.plot([0, 1], [0, 1], 'k--')  # random predictions curve
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.xlabel('False Positive Rate or (1 - Specifity)')
    plt.ylabel('True Positive Rate or (Sensitivity)')
    plt.title('Receiver Operating Characteristic')
    plt.legend(loc="lower right")
```

```python
plot_roc_curve(y, predictions)
```
