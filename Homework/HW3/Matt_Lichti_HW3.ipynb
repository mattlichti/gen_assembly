```python
import pandas as pd
import numpy as np
```

```python
student_logins = pd.read_csv('student_logins.csv', header=0)
```

```python
student_logins.info()
# student_logins.head
# student_logins.columns
```

```python
#new column duration = end_time - start_time
student_logins['duration'] = pd.to_datetime(student_logins['session_end_time']) - pd.to_datetime(student_logins['session_start_time'])
#converting duration from timedelta to float in minutes
student_logins['duration']/=np.timedelta64(1, 'm')
```

```python
student_logins['account_age'] = pd.to_datetime(student_logins['session_start_time']) - pd.to_datetime(student_logins['student_account_created'])
student_logins['account_age']/=np.timedelta64(1, 'D')
```

```python
#features is x variables, variables is x and y
features = ['problems_completed', 'student_previous_logins_total', 'student_previous_class_logins','account_age']
variables = features + ['duration']
```

```python
%matplotlib inline
from pandas.tools.plotting import scatter_matrix
scat = scatter_matrix(student_logins[variables], figsize = (10,10))
```

```python
#There are 2 outliers with duration of more than 11 hours. All other logins less than 4 hours. 
#The students probably forgot they were logged i
outliers = student_logins[student_logins['duration']>240][variables]
outliers
```

```python
#there is also 1 login with a negative duration, maybe a problem with timezones or some sort of error
huh = student_logins[student_logins['duration']<0]
huh
```

```python
#creating a new dataframe dropping the 2 outliers & negative duration
logins = student_logins[(student_logins['duration']<240) & (student_logins['duration']>0)]
```

```python
logins.info()
```

```python
scat = scatter_matrix(logins[variables], figsize = (10,10))
```

```python
y = logins['duration']
X = logins[features]
```

```python
import statsmodels.api as sm
Z = sm.add_constant(X, prepend=True)
```

```python
results = sm.OLS(y, Z).fit()
```

```python
print results.summary()
```

```python
#trying with just problems completed
X = logins['problems_completed']
Z = sm.add_constant(X, prepend=True)
results = sm.OLS(y, Z).fit()
print results.summary()
```

```python
#adding squared term
logins['prob_complete_squared'] = logins['problems_completed']**2
X = logins[['problems_completed', 'prob_complete_squared', 'student_previous_logins_total', 'student_previous_class_logins','account_age']]
Z = sm.add_constant(X, prepend=True)
results = sm.OLS(y, Z).fit()
print results.summary()
```

```python
logins['class_id'].unique()
class_dummies = pd.core.reshape.get_dummies(logins['class_id'])
class_dummies
```

## Dummy Variables

```python
logins[['class_a', 'class_c', 'class_e', 'class_g', 'class_m']]=class_dummies
```

```python
logins.columns
```

```python
#Using the 4 regular features and the 5 class dummy variables
X = logins[['problems_completed', 'student_previous_logins_total', 'student_previous_class_logins','account_age','class_a', 'class_c', 'class_e', 'class_g', 'class_m']]
Z = sm.add_constant(X, prepend=True)
results = sm.OLS(y, Z).fit()
print results.summary()
```

So far, using the best model is the one above because it has an R-squared of 0.486 meaning the 9 x variables explain 48.6% of the variation in the student login time. I could probably increase R-squared by adding interaction variables and dummy variables for the student IDs, but that could cause overfitting.
