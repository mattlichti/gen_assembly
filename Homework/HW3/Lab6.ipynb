```python
import pandas as pd
import numpy as np
```

```python
student_logins = pd.read_csv('student_logins.csv', header=0)
```

```python
student_logins.info()
# student_logins.head
# student_logins.columns
```

```python
student_logins['duration'] = pd.to_datetime(student_logins['session_end_time']) - pd.to_datetime(student_logins['session_start_time'])
# print student_logins['duration'].dtype
#converting duration from timedelta to float in minutes
student_logins['duration']/=np.timedelta64(1, 'm')
```

```python
#features is x variables, variables is x and y
features = ['problems_completed', 'student_previous_logins_total', 'student_previous_class_logins']
variables = features + ['duration']
```

```python
# %matplotlib inline
from pandas.tools.plotting import scatter_matrix
scat = scatter_matrix(student_logins[variables], figsize = (10,10))
```

```python
#There are 2 outliers with duration of more than 11 hours. All other logins less than 4 hours. 
#The students probably forgot they were logged i
outliers = student_logins[student_logins['duration']>240][variables]
outliers
```

```python
#there is also 1 login with a negative duration, maybe a problem with timezones or some sort of error
huh = student_logins[student_logins['duration']<0]
huh
```

```python
#creating a new dataframe dropping the 2 outliers & negative duration, & dropping unused columns
logins = student_logins[(student_logins['duration']<240) & (student_logins['duration']>0)][variables]
```

```python
logins.info()
```

```python
scat = scatter_matrix(logins, figsize = (10,10))
```

```python
y = logins['duration']
X = logins[features]
```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.20,random_state=0)
```

```python
import statsmodels.api as sm
Z = sm.add_constant(X, prepend=True)
```

```python
results = sm.OLS(y, Z).fit()
```
